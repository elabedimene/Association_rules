```python
from mlxtend.frequent_patterns import apriori
```

```python
#7-fonction a_priori() qui permet l’extraction des itemsets les plus fréquents. ( on définit un min_supp=0.025 et un longueur maximum de 4 produits)
frequent_itemsets = apriori(cross_tab, min_support=0.025, max_len=4, use_colnames=True)
frequent_itemsets
```

```python
#8-Afficher les 15 premiers itemsets
frequent_itemsets.head(15)
```

```python
#9-Ecrire une fonction is_inclus() qui permet de vérifier si un sous-ensemble items est inclus dans l’ensemble x.
def is_include(items , X):
  return items.issubset(X)
```

```python
#10-Afficher les itemsets comprenant le produit ‘Aspirin’.
for trs in frequent_itemsets.itemsets : 
  if trs.__contains__('Aspirin') : print(tuple(trs))
```

```python
#11-Afficher les itemsets contenant Aspirin et Eggs
for trs in frequent_itemsets.itemsets : 
   if set(('Aspirin','Eggs')).issubset(trs) :print(trs)
 
```

```python
#12-
from mlxtend.frequent_patterns import association_rules
association_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.75)
association_rules
```

```python
# 13-Afficher les 5 premières règles.
association_rules.head(5)
```

```python
# 14-Filtrer les règles en affichant celles qui présentent un LIFT supérieur ou égal à 7
association_rules[association_rules['lift'].ge(7.0)]
```

```python
# 15-Filtrer les règles en affichant celles menant au conséquent {‘2pct_milk’}
association_rules[association_rules['consequents'].eq({'2pct_Milk'})]
```
